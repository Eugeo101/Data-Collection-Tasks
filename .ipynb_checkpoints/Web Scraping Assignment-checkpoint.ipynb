```python
## Task 1
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd

response = requests.get("https://www.exchangerates.org.uk/Dollars-to-Egyptian-Pounds-currency-conversion-page.html")
soup = BeautifulSoup(response.content, 'html.parser')
type(soup)
```

```python
exchange_span = soup.find('span', attrs={'id':'shd2b;'})
us_to_eg = float(exchange_span.get_text())

while True:
    try:
        ans = float(input("Enter USD Dolars to transfer into EGP: "))
        break
    except:
        print("please enter money without any signs and numeric values only")

print(f"you entered {ans} in dollars which is {ans * us_to_eg} in EGP")
```

```python
## Task 2

response = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168")
soup = BeautifulSoup(response.content, 'html.parser')
type(soup)
```

```python
def fahrneheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5.0/9.0

names = soup.find_all('p', attrs={'class':'period-name'})
descs = soup.find_all('p', attrs={'class':'short-desc'})
temps = soup.find_all('p', attrs={'class':'temp'})

def to_str(arg):
    return str(arg.text)

def to_celisuis(arg):
    return fahrneheit_to_celsius(float(str(arg.text).split(' ')[1]))
    
names = list(map(to_str, names))
descs = list(map(to_str, descs))
temps = list(map(to_celisuis, temps))

with open('temps.csv', 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['name', 'description', 'temprature'])
    writer.writeheader()
    for i in range(len(names)):
        name = names[i]
        desc = descs[i]
        temp = temps[i]
        writer.writerow({'name':name, 'description':desc, 'temprature':temp})

df = pd.read_csv('temps.csv')
df
```

```python
## Task 3

response = requests.get('https://books.toscrape.com/catalogue/page-1.html')
soup = BeautifulSoup(response.content, 'html.parser')
number_of_pages = int(str(soup.find('li', attrs={'class':'current'}).text).strip().split(' ')[-1])
star_to_rate = {'One':1, 'Two':2, 'Three':3, 'Four':4, 'Five':5}
number_of_pages
```

```python
with open('books.csv', 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['name', 'star', 'price', 'is_available'])
    writer.writeheader()
    
    for page in range(1, number_of_pages+1):
        print(f"Scrapping page {page}")
        page = f'https://books.toscrape.com/catalogue/page-{page}.html'
        response = requests.get(page)
        soup = BeautifulSoup(response.content, 'html.parser')
        articles = soup.find_all('article', attrs={'class':'product_pod'})
        for article in articles:    
            name = article.find('img').attrs['alt']
            star = star_to_rate[str(article.find('p').attrs['class'][-1])]
            price = float(str(article.find('p', attrs={'class':'price_color'}).text)[1:])
            is_available = article.find('p', attrs={'class':'instock availability'}).find('i').attrs['class'][0] == 'icon-ok'
            writer.writerow({'name':name, 'star':star, 'price':price, 'is_available':is_available})
```

```python
import chardet
char_det = chardet.detect(open('books.csv', 'rb').read())
df = pd.read_csv('books.csv', encoding=char_det['encoding'])
df.sort_values(by='star', ascending=False)
```

```python
## Task 4

# scrape the 250 countries (name, capital, population, area) and save them into a csv file 
response = requests.get('https://www.scrapethissite.com/pages/simple/')
soup = BeautifulSoup(response.content, 'html.parser')
type(soup)
```

```python
names = soup.find_all('h3', attrs={'class':'country-name'})
capitals = soup.find_all('span', attrs={'class':'country-capital'})
populations = soup.find_all('span', attrs={'class':'country-population'})
areas = soup.find_all('span', attrs={'class':'country-area'})

with open('countries_info.csv', 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['name', 'capital', 'population', 'area'])
    for i in range(len(names)):
        name = names[i].text.strip()
        capital = capitals[i].text.strip()
        population = populations[i].text.strip()
        area = areas[i].text.strip()
        try:
            writer.writerow({'name':name, 'capital':capital, 'population':population, 'area':area})
        except:
            print(names[i], capitals[i], populations[i], areas[i])
            pass
        

result = chardet.detect(open('countries_info.csv', 'rb').read())
df = pd.read_csv('countries_info.csv', encoding=result['encoding'])
df
```

```python
## Task 5

# Team name, year, wins, losses, win%, goals for(gf), golaagainst(ga)
response = requests.get('https://www.scrapethissite.com/pages/forms/?page_num=1')
soup = BeautifulSoup(response.content, 'html.parser')
type(soup)
```

```python
with open('teams_info.csv', 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['Team Name', 'year', 'wins', 'losses', 'win%', 'goals for(gf)', 'golaagainst(ga)'])
    writer.writeheader()
    all_teams = soup.find_all('tr', attrs={'class':'team'})
    for team in all_teams:
        name = team.find('td', attrs={'class':'name'}).get_text()
        year = team.find('td', attrs={'class':'year'}).get_text()
        wins = team.find('td', attrs={'class':'wins'}).get_text()
        losses = team.find('td', attrs={'class':'losses'}).get_text()
        pct = team.find('td', attrs={'class':'pct'}).get_text()
        gf = team.find('td', attrs={'class':'gf'}).get_text()
        ga = team.find('td', attrs={'class':'ga'}).get_text()
        writer.writerow({'Team Name':name, 'year':year, 'wins':wins, 'losses':losses, 'win%':pct, 'goals for(gf)':gf, 'golaagainst(ga)':ga})
        

result = chardet.detect(open('teams_info.csv', 'rb').read())
df = pd.read_csv('teams_info.csv', encoding=result['encoding'])
df
```

```python
## Task 6

# name, username, email, street, suite, city, zip code, geo_lat & geo_long
response = requests.get('https://jsonplaceholder.typicode.com/users')
users = response.json()
len(users)
```

```python
with open('users.csv', 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['name', 'username', 'email', 'street', 'suite', 'city', 'zip code', 'latitude', 'longitude'])
    writer.writeheader()
    for user in users:
        name = user['name']
        user_name = user['username']
        email = user['email']
        street = user['address']['street']
        suite = user['address']['suite']
        city = user['address']['city']
        zip_code = user['address']['zipcode']
        geo_lat = user['address']['geo']['lat']
        geo_lng = user['address']['geo']['lng']
        writer.writerow({'name':name, 'username':user_name, 'email':email, 'street':street, 'suite':suite, 'city':city, 'zip code':zip_code, 'latitude':geo_lat, 'longitude':geo_lng})

df = pd.read_csv('users.csv')
df
```
